```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def imshow(im):
    width, height, *channels = im.shape
    if channels:
        # By default, OpenCV tends to work with images in the BGR format.
        # This is due to some outdated practices, but it has been left in the library.
        # We can iterate the channels in reverse order to get an RGB image.
        plt.imshow(im[:,:,::-1])
    else:
        plt.imshow(im, cmap='gray')
    # plt.axis('off')
    plt.show()

```

```python

def cvv2_gen():
    return str(np.random.randint(1000, 9999))

def date_gen():
    month = np.random.randint(0,13)
    if month < 10:
        month = '0' + str(month)
    else:
        month = str(month)
    year = '0' + str(np.random.randint(0, 10))
    return f'{year}/{month}'

def num_gen():
    first = str(np.random.randint(5000, 6400))
    second = str(np.random.randint(600, 9999))
    second = '0'*(4 - len(second)) + second
    third = str(np.random.randint(1, 9999))
    third = '0'*(4 - len(third)) + third
    forth = str(np.random.randint(1, 9999))
    forth = '0'*(4 - len(forth)) + forth
    return first + ' ' + second + ' ' + third + ' ' + forth

def get_color(black=False):
    if not black:
        return (np.random.randint(0, 200), np.random.randint(0, 200), np.random.randint(0, 200))
    return (np.random.randint(0, 10), np.random.randint(0, 10), np.random.randint(0, 10))
```

```python

img = cv2.imread('./meli_1.jpg')
bank_dir = './melly/'

height, width = img.shape[0:2]

for i in range(100, 160):
    im = img.copy()
    date = date_gen()
    cvv2 = cvv2_gen()
    num = num_gen()

    #cvv2
    color = (np.random.randint(0, 200), np.random.randint(0, 200), np.random.randint(0, 200))
    xc, yc = 50, 340
    cv2.putText(im, 'CVV2:'+cvv2, (xc, yc), 1, 1.5, color, 4)
    a, b = 20, 130
    pt1 = (xc, yc - a)
    pt2 = (xc + b, yc)
    wc = pt2[0] - pt1[0]
    hc = pt2[1] - pt1[1]
    # cv2.rectangle(im, pt1, pt2, (0,0,255), 2)

    x_center_c = float((2*xc+b)//2)/width
    y_center_c = float((2*yc-a)//2)/height
    w_c = float(wc)/width
    h_c = float(hc)/height

    # cr num
    color = (np.random.randint(0, 200), np.random.randint(0, 200), np.random.randint(0, 200))
    xn, yn = 60, 275
    cv2.putText(im, num, (xn, yn), 1, 2.7, color, 4)
    a, b = 32, 495
    pt1 = (xn, yn - a)
    pt2 = (xn + b, yn)
    wn = pt2[0] - pt1[0]
    hn = pt2[1] - pt1[1]
    # cv2.rectangle(im, pt1, pt2, (0,0,255), 2)

    x_center_n = float((2*xn+b)//2)/width
    y_center_n = float((2*yn-a)//2)/height
    w_n = float(wn)/width
    h_n = float(hn)/height


    with open(bank_dir + f'{i}.txt', 'w') as f:
        f.write(f'0 {x_center_n} {y_center_n} {w_n} {h_n}\n')
        f.write(f'1 {x_center_c} {y_center_c} {w_c} {h_c}')
    cv2.imwrite(bank_dir + f'{i}.jpg', im)
imshow(im)

# cv2.imwrite('mellat.png', img)
```
